```python
import pandas as pd

# Ler o arquivo CSV com os dados de chuva diária
dados = pd.read_excel('ilha bela.xlsx')
```

```python
display(dados)
```

```python
dados.head(15)
```

```python
dados = dados.drop(index=range(0, 9), axis=0)
dados.head(15)
```

```python
novo_cabecalho = dados.loc[9]
dados = dados.set_axis(novo_cabecalho, axis=1)
dados.head(15)
```

```python
dados_VF = dados.drop(index=9, axis=0)
display(dados_VF)
```

```python
dados_VF = dados_VF.dropna()
display(dados_VF)
```

```python
dados_VF.iloc[:, 0] = pd.to_datetime(dados_VF.iloc[:, 0], format='%Y-%m-%d')
```

```python
dados_VF['dia'] = dados_VF.iloc[:, 0].dt.day
dados_VF['mes'] = dados_VF.iloc[:, 0].dt.month
dados_VF['ano'] = dados_VF.iloc[:, 0].dt.year
```

```python
display(dados_VF)
```

```python
# Calcular a média diária de chuva para cada dia e mês
media_diaria_dia = dados_VF.groupby(['dia'])['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'].mean()
media_diaria_mes = dados_VF.groupby(['mes'])['PRECIPITACAO TOTAL, DIARIO (AUT)(mm)'].mean()

# Encontrar o dia e o mês com a menor média de chuva
dia_min_chuva = media_diaria_dia.idxmin()
mes_min_chuva = media_diaria_mes.idxmin()

print(f"O dia com menor média de chuva em 2024 é {dia_min_chuva} e o mês com menor média de chuva é {mes_min_chuva}")
```
